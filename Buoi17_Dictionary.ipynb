```python
# Thông tin phòng ban
dept_db = {
    101 : 'HRD',
    102 : 'FINANCE',
    103 : 'ACCOUNTS',
    104 : 'SALES',
    105 : 'ENGINEERING',
    106 : 'SUPPORT'
}

employee_db = { 
    1000: {
        "name" : "Alex",
        "doj" : '01-10-89',
        "dept" : 103
    },
    1001: dict(name="Mary",   doj='01-11-88', dept=101),
    1002: dict(name="John",   doj='01-10-87', dept=104),
    1003: dict(name="David",  doj='01-06-89', dept=105),
    1004: dict(name="Anne",   doj='01-01-86', dept=106),
    1005: dict(name="Samson", doj='01-02-89', dept=101)
}

print(employee_db)
print()

# Liệt kê danh sách nhân viên của phòng dept=101
for key in employee_db:
    if employee_db[key]["dept"] == 101:
        print(employee_db[key])
        
# Dùng hàm filter + lambda để lọc các phần của list/dict
result = filter(lambda item: employee_db[item]["dept"] == 101, employee_db)
print(list(result))
```

```python
"""Thống kê số từ của đoạn văn bản nhập vào."""
doan_van = """
Hàm sẽ nhận tham số truyền vào là chuỗi văn bản cần đếm
Trong hàm, khai báo biến num_words, mang giá trị là các cặp key-value, tương ứng với từ được dếm và số lần xuất hiện của từ đó
Sử dụng hàm split để tách các từ trong chuỗi truyền vào và lưu thành một mảng, gán giá trị này cho biến text_list
Sử dụng for…in để lặp qua mảng text_list, giá trị lặp sẽ là biến từ
Sử dụng lower để chuyển các ký tự hoa trong từ thành ký tự thường, sử dụng strip() để loại bỏ các khoản trắng đằng trước hoặc đằng sau từ (nếu có), sau đó gán giá trị vừa được cập này cho từ đang lặp.
Sử dụng hàm get để lấy giá trị đếm hiện tại của từ lặp trong num_words, nếu không tìm thấy thì giá trị mặc định ban đầu sẽ là 0
Sau khi lấy được giá trị đếm của từ lặp trong num_words, cộng giá trị thêm 1 và gán giá trị này cho biến num_words[word] (câu lệnh này cũng sẽ thực hiện việc thêm key-value cho num_words nếu không tìm thấy từ lặp trong num_words)
Trả biến num_words làm kết quả hàm
"""

doan_van = doan_van.lower()
# print(doan_van)

# bóc tách các từ dựa vào khoảng trắng
mang_tu = doan_van.split()
# print(mang_tu)
# print()

# Xóa những kí tự trùng lắp trong
# Ý tưởng: chuyển list() thành dict() với key là list() đó, sau đó ép kiểu list
mylist = list(dict.fromkeys(mang_tu))
# print(mylist)
for tu in mylist:
    print(tu, ':', doan_van.count(tu) ,'lần')
```
